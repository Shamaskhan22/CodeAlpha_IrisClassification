# Task 2: Unemployment Analysis with Python

● Analyze unemployment rate data representing unemployed people percentage

```python
import pandas as pd
data = pd.read_csv("Unemployment in India.csv")  
print(data.head())
```

● Use Python for data cleaning, exploration, and visualization of unemployment trends.

```python
print(data.info())
print(data.describe())
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data.columns = data.columns.str.strip()


plt.figure(figsize=(12,6))
sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", data=data, marker="o")
plt.title("Unemployment Rate Over Time in India")
plt.xticks(rotation=45)
plt.show()

# Covid-specific trend (2020-2021 data)
covid_data = data[(data["Date"].str.contains("2020")) | (data["Date"].str.contains("2021"))]

plt.figure(figsize=(12,6))
sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", data=covid_data, color="red", marker="o")
plt.title("Unemployment Rate During Covid-19 (2020-2021)")
plt.xticks(rotation=45)
plt.show()
```

● Identify key patterns or seasonal trends in the data.

```python
# Group by year if year column available
if "Year" in data.columns:
    yearly = data.groupby("Year")["Unemployment Rate"].mean()
    print(yearly)
    yearly.plot(kind="bar", figsize=(8,5), title="Yearly Average Unemployment Rate")
    plt.show()
```
